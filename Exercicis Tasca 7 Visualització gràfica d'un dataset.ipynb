# Visualització gràfica d'un dataset

## Exercici 1

```python
# imports

import pandas as pd 
import numpy as np
import seaborn as sns 
from os import path
from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator

import matplotlib.pyplot as plt
```

```python
dataframe= pd.read_csv('C:/Users/TREBALL/Desktop/BCN-Activa-Data Science/database/DelayedFlights.csv')
df = pd.DataFrame(dataframe)
```

### Visualització per a variable categòrica

```python
plt.plot(df['UniqueCarrier'])
plt.savefig('C:/Users/TREBALL/Desktop/BCN-Activa-Data Science/database/catego.png')
```

###  Visualització per a variable numèrica

```python
plt.plot(df['ArrDelay'])
plt.savefig("Visualització per a variable numèrica.jpg", bbox_inches='tight')
plt.savefig('C:/Users/TREBALL/Desktop/BCN-Activa-Data Science/database/nume.png')
```

### Visualització per a variable numèrica i una categòrica

```python
plot_order = df.groupby('UniqueCarrier')['ArrDelay'].sum().sort_values(ascending=False).index.values

Aerol_delay = df[['UniqueCarrier', 'ArrDelay']]
Aerol_delay = Aerol_delay.groupby(by='UniqueCarrier').sum()
Aerol_delay=Aerol_delay.reset_index(drop=False)
Aerol_delay.head(7)
plt.figure(figsize=(10,6))
sns.barplot(x="UniqueCarrier", y="ArrDelay", data=Aerol_delay, order=plot_order[1:15])
plt.show(10)

plt.savefig('C:/Users/TREBALL/Desktop/BCN-Activa-Data Science/database/numeicatego.jpg', dpi=300)
```

### Visualització per a dues variables numèriques 

```python
plot_order = df.groupby('DepDelay')['ArrDelay'].sum().sort_values(ascending=False).index.values

Aerol_delay = df[['DepDelay', 'ArrDelay']]
Aerol_delay = Aerol_delay.groupby(by='DepDelay').sum()
Aerol_delay=Aerol_delay.reset_index(drop=False)
Aerol_delay.head(7)
plt.figure(figsize=(10,6))
sns.barplot(x="DepDelay", y="ArrDelay", data=Aerol_delay, order=plot_order[1:15])
plt.show(10)

plt.savefig("C:/Users/TREBALL/Desktop/BCN-Activa-Data Science/database/dosnumeriques.jpg")
```

### Visualització per a tres variables

```python
x=df['UniqueCarrier']
y1=df['ArrDelay']
y2=df['DepDelay']

plt.stackplot(x,y1, y2, labels=['ArrDelay','DepDelay'])
plt.legend(loc='upper left')
plt.savefig("C:/Users/TREBALL/Desktop/BCN-Activa-Data Science/database/tresvariables.jpg", bbox_inches='tight')
```

### Visualització per a més de tres variables

```python
x=df['UniqueCarrier']
y1=df['ArrDelay']
y2=df['DepDelay']
y3=df['AirTime']

# Basic stacked area chart.
plt.stackplot(x,y1, y2, y3, labels=['ArrDelay','DepDelay','AirTime'])
plt.legend(loc='upper left')
plt.savefig("C:/Users/TREBALL/Desktop/BCN-Activa-Data Science/database/mesdtresvariables.jpg", bbox_inches='tight')
```
